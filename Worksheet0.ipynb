```python
#7
#Ishan Khokhali
# Import library for plotting
import matplotlib.pyplot as plt

# Temperature data (in Â°C)
temperatures = [
    8.2, 17.4, 14.1, 7.9, 18.0, 13.5, 9.0, 17.8, 13.0, 8.5,
    16.5, 12.9, 7.7, 17.2, 13.3, 8.4, 16.7, 14.0, 9.5, 18.3,
    13.4, 8.1, 17.9, 14.2, 7.6, 17.0, 12.8, 8.0, 16.8, 13.7,
    7.8, 17.5, 13.6, 8.7, 17.1, 13.8, 9.2, 18.1, 13.9, 8.3,
    16.4, 12.7, 8.9, 18.2, 13.1, 7.8, 16.6, 12.5
]

# Task 1: Categorize temperatures into cold, mild, and comfortable
cold = []        # Below 10Â°C
mild = []        # 10Â°C to <15Â°C
comfortable = [] # 15Â°C to 20Â°C

for temp in temperatures:
    if temp < 10:
        cold.append(temp)
    elif 10 <= temp < 15:
        mild.append(temp)
    else:
        comfortable.append(temp)

# Print categorized lists
print("Cold temperatures:", cold)
print("Mild temperatures:", mild)
print("Comfortable temperatures:", comfortable)

# Task 2: Count how many temperatures are in each category
print("\nTask 2 Results:")
print(f"Cold readings: {len(cold)}")
print(f"Mild readings: {len(mild)}")
print(f"Comfortable readings: {len(comfortable)}")

# Task 3: Convert temperatures to Fahrenheit
temperatures_fahrenheit = [(temp * 9/5) + 32 for temp in temperatures]
print("\nTemperatures in Fahrenheit:")
print(temperatures_fahrenheit)

# Task 4: Analyze temperatures by time of day
night = []   # Night readings
evening = [] # Evening readings
day = []     # Day readings

for i in range(len(temperatures)):
    if i % 3 == 0:       # Night
        night.append(temperatures[i])
    elif i % 3 == 1:     # Evening
        evening.append(temperatures[i])
    else:                # Day
        day.append(temperatures[i])

# Calculate average daytime temperature
average_day_temp = sum(day) / len(day)
print("\nTask 4 Results:")
print(f"Average daytime temperature: {average_day_temp:.2f}Â°C")

# Plot daytime temperatures
plt.figure(figsize=(10, 5))
plt.plot(range(1, len(day) + 1), day, marker='o', linestyle='-', color='orange', label='Day Temperatures')
plt.title("Day vs. Temperature")
plt.xlabel("Day")
plt.ylabel("Temperature (Â°C)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#8
#Ishan Khokhali
# Task 1: Calculate the sum of all numbers in a nested list

def sum_nested_list(nested_list):
    """
    Sum all numbers in a nested list.
    """
    total = 0
    for element in nested_list:
        if isinstance(element, list):  # If the element is a list, call function recursively
            total += sum_nested_list(element)
        else:  # Otherwise, add the number directly
            total += element
    return total

# Test cases
nested_list_1 = [1, [2, [3, 4], 5], 6, [7, 8]]
nested_list_2 = [1, [2, 3], [4, [5]]]

print(f"Sum of {nested_list_1}: {sum_nested_list(nested_list_1)}")  # Output: 36
print(f"Sum of {nested_list_2}: {sum_nested_list(nested_list_2)}")  # Output: 15

# Task 2: Generate all unique permutations of a string

def generate_permutations(s):
    """
    Generate unique permutations of a string.
    """
    if len(s) == 1:  # Base case: single character
        return [s]
    
    permutations = []
    for i, char in enumerate(s):
        # Get the remaining string excluding the current character
        remaining = s[:i] + s[i+1:]
        for perm in generate_permutations(remaining):  # Recursive call
            new_perm = char + perm
            if new_perm not in permutations:  # Avoid duplicates
                permutations.append(new_perm)
    return permutations

# Test cases
print(f"Permutations of 'abc': {generate_permutations('abc')}")  # ['abc', 'acb', 'bac', 'bca', 'cab', 'cba']
print(f"Permutations of 'aab': {generate_permutations('aab')}")  # ['aab', 'aba', 'baa']

# Task 3: Calculate the total size of a directory

def calculate_directory_size(directory):
    """
    Calculate the total size of a directory.
    """
    total_size = 0
    for key, value in directory.items():
        if isinstance(value, dict):  # If it's a subdirectory, calculate its size
            total_size += calculate_directory_size(value)
        else:  # Otherwise, add file size
            total_size += value
    return total_size

# Sample directory structure
directory_structure = {
    "file1.txt": 200,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    },
    "subdir2": {
        "subsubdir1": {
            "file5.txt": 250
        },
        "file6.txt": 150
    }
}

# Test case
print(f"Total directory size: {calculate_directory_size(directory_structure)} KB")  # Output: 1400 KB
```

```python
#Ishan Khokhali
# Task 1: Minimum Coins
def min_coins(coins, amount):
    """
    Find the minimum number of coins needed to make a given amount.
    Returns -1 if the amount cannot be made.
    """
    dp = [float('inf')] * (amount + 1)  # Initialize DP array with infinity
    dp[0] = 0  # Base case: 0 coins to make amount 0

    for coin in coins:
        for i in range(coin, amount + 1):  # Update DP array for each coin
            dp[i] = min(dp[i], dp[i - coin] + 1)

    return dp[amount] if dp[amount] != float('inf') else -1

# Example usage
print(min_coins([1, 2, 5], 11))  # Output: 3


# Task 2: Longest Common Subsequence
def longest_common_subsequence(s1, s2):
    """
    Find the length of the longest common subsequence of two strings.
    """
    length1, length2 = len(s1), len(s2)
    dp = [[0] * (length2 + 1) for _ in range(length1 + 1)]  # DP table for lengths

    for i in range(1, length1 + 1):
        for j in range(1, length2 + 1):
            if s1[i - 1] == s2[j - 1]:  # Characters match
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:  # Take the maximum from previous states
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

    return dp[length1][length2]

# Example usage
print(longest_common_subsequence("abcde", "ace"))  # Output: 3


# Task 3: Knapsack Problem
def knapsack(weights, values, capacity):
    """
    Solve the 0/1 Knapsack problem to find the maximum value for a given capacity.
    """
    dp = [[0] * (capacity + 1) for _ in range(len(weights) + 1)]  # DP table

    for i in range(1, len(weights) + 1):
        for w in range(1, capacity + 1):
            if weights[i - 1] <= w:  # Item can be included
                dp[i][w] = max(dp[i - 1][w], values[i - 1] + dp[i - 1][w - weights[i - 1]])
            else:  # Item cannot be included
                dp[i][w] = dp[i - 1][w]

    return dp[len(weights)][capacity]

# Example usage
weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7
print(knapsack(weights, values, capacity))  # Output: 9
```
